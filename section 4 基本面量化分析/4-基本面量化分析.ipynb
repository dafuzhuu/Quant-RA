# 1.基本面量化简介

## 1.1 基本面量化背景
### 1.1.1 基本面量化崛起的时代背景
  * 欧美投资界在围绕数理金融的纯量化交易中发现：数理金融的抽象理论逐渐弱化，金融市场本身的逻辑愈加重要。纯市场交易的数量模型存在单一性和有效性不足的缺点。
  * 近年来，因子投资（factor investing）被广泛接受。其中很多因子是通过财务报表中的会计学指标构造的，比如 Book-to-Market Value、Return on Equity 等，它们被称为基本面因子。定量使用这类因子进行的基本面投资则被称为基本面量化投资（Quantamental）。
  * 量化基本面分析法作为典型的主动数量化投资，越来越受到国内公私募基金管理人的认可。### 1.1.2 华尔街+学术界联合推动基本面量化崛起

* 量化基本面基金大多基于基本面变量+资产组合技术，开发出大量的选股策略；* 因子投资的普及造就了数不胜数的 smart beta ETFs 产品，其中不乏对标基本面分析的 Value和 Quality 因子类产品。低成本 ETFs 让普通投资者享受因子投资的同时，也推波助澜了基本面分析的“因子化”;* 面对如此影响深远的变化，Wall Street Journal 不禁高呼 “The Quants Run Wall Street Now”。<p></p>

### 1.1.4 基本面分析
* 基本面分析源自 Graham and Dodd (1934) 这本著名的 Security Analysis；基本面分析旨在通过定量和定性分析与上市公司相关的经济和金融数据来衡量证券的内在价值。而 SecurityAnalysis 所提倡的基本面分析技术直接催生了金融分析师（financial analysts）这个职业。
* 1947 年，美国一些地方的证券分析师协会合并，成立了第一个全国性的金融分析师协会（National Federation of Financial Analysts Societies）。随后，该协会更名为 Financial Analysts Federation（FAF）；
* 1959 年，Institute of Chartered Financial Analysts（ICFA）成立并开始举办 Chartered Financial Analysts（CFA）考试。最终，FAF 和 ICFA 合并组成了如今家喻户晓的 CFA Institute

### 1.1.5 有效市场理论的挑战
* 自上世纪 50 年代开始，现代投资组合理论、CAPM、有效市场假说（EMH）以及期权定价相继被提出，奠定了金融学的基础。而它们之中，对基本面分析造成最大冲击的当属有效市场假说。
* Fama (1965) 写到：“If the random walk theory is valid and if security exchanges are ‘efficient 'markets, then stock prices at any point in time will represent good estimates of intrinsic or fundamental values. Thus, additional fundamental analysis is only of value when the analyst has new information which was not considered in forming current market prices or has new insights concerning the effects of generally available information which are not already implicit in current prices.”

* 如果市场是有效的，那么价格已经很好的反映了证券的内在价值。因此，除非分析师有新的消息（或者对已有消息有新的解读），否则是无法通过已有财报找到定价错误的。虽然 EMH从推出以来就一直伴随着争议（往回翻仔细看本文第一张图左上角 Buffett 的那句话），但它依然作为市场第一范式牢牢统治了学术界的课堂。

## 1.2 基本面量化发展历程
* 对于股票市场的定量分析很大程度上得益于上世纪 60 年代 CRSP 推出了股票价格数据，因为它让回测成为可能。有了数据的支持，学术界各显神通，先是对 CAPM 进行了检验并拒绝了它，而后又接二连三的发现了一个又一个的异象（anomalies）或因子（factor）。比如著名的Fama and French (1993) 三因子中的 Book-to-Market Ratio（B/M）就是一个 value 因子。

* 随着学术界在挖掘异象的道路上越走越远，业界自然也没闲着。业界把学术界的研究成果很好的落地化，变成了许多 style indices（风格指数）。举例来说，1987 年 Russell  Investments推出了最早的两个风格指数，分别对标价值股和成长股，而构造方法恰恰是 B/M：B/M 前一半的股票被放在了 Value Index、B/M 后一半的股票被放在了 Growth Index。在这之后，S&P、MSCI 以及 CRSP 也推出了它们各自的风格指数。### 1.2.1 Index-based Value Funds（指数基金）
* 最初，推出这些指数的目的是用来评估主动基金经理人的业绩。然而不久之后，复制这些指数的基金便应运而生，而这其中的代表则是大名鼎鼎的 Vanguard。1992 年，Vanguard 推出了第一支 value index fund（和第一支 growth index fund）：“The Value Portfolio seeks toreplicated the aggregate price and yield performance of the S&P/Barra Value Index, an index which includes stocks in the S&P 500 Index with lower than average ratios of market prices to book value. These types of stocks are often referred to as ‘value’ stocks.”
* Index-based Value Funds 一经推出便受到了追捧。人们把它们视作主动型价值投资的低成本替代品。鉴于 Value 因子取得的巨大成功，业界也开始把关注的重点进而转移到学术界发现的其他显著异象，并构造出了一系列 smart beta 产品。在这之中，除了 value 之外另一个高度依赖财务数据的则是 quality 因子。### 1.2.2 ETF基金的推出
* 下图展示了 MSCI、Russell 以及 Fidelity 三家提供的 quality index 的选股标准。颇有意思的是这三个构造方法都是用了某种投资回报率指标（如 ROE 或 ROA），很大程度上得益于学术界在 profitability 上的研究成果；反映出业界对于学术界的最新发现追踪的非常及时。截止 2018 年 4 月，规模最大的 quality 因子 ETF—— the iShares Edge MSCI USA Quality Factor ETF —— 的资金量高达 45 亿美元。

## 1.3 研究文献：基本面分析研究方法
* 大量研究基于美国市场数据发现利用公司基本面特征构建不同类型因子，可以解释或预测不同股票在时间序列和横截面上的收益差异。
* Green et al.（2017）研究了94个公司特征的预测能力，Yan and Zheng（2017）使用自助法（bootstrap）从财务报表中构建了大量信号用于预测横截面收益。至今为止，学术界宣称发现了数以百计能够提供超额收益的异象（因子），前美国金融学会主席Cochrane（2022）称之为“因子动物园”（Factor Zoo），然而后续的样本外检验却发现大部分因子难以持续地提供超额收益（Hou et al.，2019）。
* 近年来学者开始探索运用机器学习方法来解决因大量因子涌现带来的技术挑战，主要研究思路有如下三种：

  * 采用变量选择方法衡量因子对于资产定价的贡献，比如Feng et al.（2017）采用Lasso方法来衡量因子对资产定价的贡献。
  * 运用机器学习方法从因子中提取共同因素来解释横截面收益差异，比如Light et al.（2017）采用“偏最小二乘法”（Partial Least Square，PLS）来检验公司基本面特征对期望收益的预测能力；Kozak et al.（2019）和Kelly et al.（2020）分别运用PCA方法和IPCA方法（Instrumented PCA）提取因子中的共同因素。
  * 设计新的集成方法以获取更好的预测效果，比如Lewellen（2015）发现通过Fama-French回归方法综合15个因子能够很好地预测股票超额收益。 ## 1.4 研究文献：A股基本面分析的有效性
* 胡熠和顾明（2018）从安全性、便宜性和质量三个维度选取企业特征构造综合性指标B-score，发现B-score对股票未来收益有很好的预测能力。* 李斌等（2019）运用多种机器学习算法提取A股市场的96项异象因子构建股票收益预测模型及投资组合，证实机器学习算法能够有效地识别异象因子与超额收益之间的复杂关系。* Jiang et al.（2019）采用多种大数据降维方法整合A股市场中的75个公司特征信息，发现交易摩擦、动量和盈利能力相关的公司特征是A股收益有效的预测因子。* 姜富伟等（2022）利用多种机器学习算法对高维财务基本面数据进行降维和变量选择。* Leippold et al.（2022）利用多种机器学习算法全面分析了A股上市企业收益的预测因子，发现流动性是A股最重要的预测因子。

# 2.行业基本面量化
## 2.1 行业基本面量化的定义和投资逻辑
* <b>定义：</b>所谓行业基本面量化，主要就是在股票投资中，以基本面量化的手段，选取未来能够显著跑赢市场的行业。多数情况下，股票市场对处于高度景气周期中的行业资产已经进行了相对合理的“估价”。因此行业基本面量化主要的目标，在于寻找到<b>高效的先行因子，</b>来预判未来哪些行业在利润等主要基本面指标上，能够显著领先于其他多数行业。<p></p>
* <b>基本逻辑：</b>行业资产配置的一个基本逻辑是，各行业在相同的经济环境却可能遵循着不同的经济模式。一方面，经济内在发展本身所具有的行业特点差异所致，有的行业表现好的时候意味着其它行业表现差，如上下游行业之间；另一方面则是大量情况证实，不同行业的股票在市场定价中，往往会表现出鲜明的行业轮动特点。

## 2.2 搭建备选因子库
* 在行业基本面的量化因子选择方面，实证性研究主要的出发点，不是瞄准当前行业的基本面情况，而是为了对未来行业整体景气度和利润状况进行预判，确定备选的基本面量化因子库。<p></p>

* 考虑到中国经济周期的特殊性，以及近年来愈发明显地对货币环境的依赖性，行业基本面的先行因子，大致可以划分为三大类：宏观因子、货币因子和行业因子。

### 3.2.3 因子筛选及模型构建及结论
* <b>回归法：</b>使用股票过去的收益率对基本面因子的大类进行回归，然后得到一个线性回归方程。在该回归方程回测有效和适当修正之后，再把最新的因子值带入回归方程得到一个对未来股票收益的预判。
  * 例：  
    中国上市公司研究院对截至2016年5月3日（2015年年报披露结束后首个交易日）的样本数据所做的因子分析显示：将公因子作为自变量与股票价格P做回归分析，影响股票价格的主要因子有5个：  
    <b>P = 19.6135 + 0.6207F1 +1.6508F2 + 5.845F3 – 1.6475F4 + 0.8037F7</b>  
    其中，F1为成长因子，F2为偿债因子，F3为盈利因子，F4为规模因子，F7为运营因子<p></p>

<b>不同类型因子的系数大小，直观反映了其对于股价影响程度的高低。</b>
* <b>打分法：</b>根据各个因子值进行排序，将其在所有股票因子值的相对位置给予该个股一个分数，再对个股的所有因子值进行加权平均得到该股票的最终得分。最后将所有个股的综合得分进行排序，挑选出分数较高的股票。
  * 公式：
    <center>
        \begin{multline*}
        基本面因子得分 = \sum_ {}{}{a\times基本面类因子得分} （其中a为因子权重）
        \end{multline*}
    </center>
  * 优：打分法相对较稳健，不容易受到极端值的影响
  * 劣：因各个因子的权重需要人为设定，主观性也较强
  * <b>评价打分法模型的有效性：</b>  
    先利用Sharpe比率（或其他常用的投资绩效评价指标）进行有效因子的筛选通过，再构造打分方法，进而在样本股票内检验该打分方法的有效性，确认有效之后再在样本外测试该打分法选股的收益情况。往复统计检验之后，最终确定最有效的因子类型和打分模型<p></p>

## 3.3 Python案例代码

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# 假设我们有以下几个公司的数据和一些基本面指标
data = {
    'company': ['A', 'B', 'C', 'D', 'E'],
    'pe_ratio': [15, 20, 12, 18, 16],
    'eps_growth': [0.1, 0.15, 0.08, 0.12, 0.09],
    'roa': [0.12, 0.1, 0.08, 0.11, 0.09],
    'debt_to_equity': [0.5, 0.6, 0.7, 0.4, 0.55],
    'stock_return': [0.2, 0.15, 0.1, 0.18, 0.12]
}

df = pd.DataFrame(data)

# 定义可能的基本面因子
factors = ['pe_ratio', 'eps_growth', 'roa', 'debt_to_equity']

# 通过回归法进行因子筛选
selected_factors = []
for factor in factors:
    X = df[factor].values.reshape(-1, 1)
    y = df['stock_return'].values
    model = LinearRegression()
    model.fit(X, y)
    if model.score(X, y) > 0.5:  # 这里设置一个简单的阈值，可根据实际情况调整
        selected_factors.append(factor)
        
print(f"Selected factors: {selected_factors}")

# 通过打分法进行股票选择
# 设定筛选后指标的权重
weights = {factor: 1 / len(selected_factors) for factor in selected_factors}

# 定义指标评分函数
def score_feature(feature, ascending=False):
    sorted_vals = np.sort(df[feature])
    scores = np.linspace(0, 100, len(sorted_vals))
    if ascending:
        return pd.Series([scores[np.where(sorted_vals == val)[0][0]] for val in df[feature]], index=df.index)

# 对每个筛选后的指标进行评分
scored_features = {}
for feature, weight in weights.items():
    print(feature,weight)
    df[f'{feature}_score'] = score_feature(feature, ascending=ascending)

# 计算综合得分

df['total_score'] = 0
for feature, weight in weights.items():
    df['total_score'] += df[f'{feature}_score'] * weight

print(df)
```

# 总结
* 量化基本面分析遵循top-down，即自上而下的分析模式，从国家总体形势到行业，再到个股。* 行业的基本面量化是核心，如何捕捉板块和相关概念往往决定了大部分收益，而行业量化的核心在于如何发现前瞻性指标，其来源往往是对政策的分析和市场的敏感。* 个股的基本面量化其实效果不一定好，主要原因在于个股层面往往存在非常多的内幕信息，市场能够获取的公开信息往往已经滞后于市场价格，因此个股层面的量化主要还是在于按照一定的准则去掉基本面差的股票。* 基本面量化可以基于复杂模型，也可以基于简单的打分法，复杂的模型带来的收益并不一定强过简单的分析方法，其关键还是在于有价值的数据。
