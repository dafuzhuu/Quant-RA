# 一、商品期货基金（CTA) 简介
## 1.1 策略原理
CTA 全称是 commodity trading advisor，是提供有关期货合约、期货期权和零售场外外汇合约或掉期买卖个性化建议的个人或公司。  
CTA 策略， 又称“ 管理期货”（Managed Futures），泛指**以期货为主要交易标的的投资策略或资产管理产品**。

- 分析方法：主观CTA、量化CTA
- 交易策略：趋势跟踪策略、反转策略、套利策略
- 交易频率：日内高频、短周期、中周期、长周期

CTA 策略基金在总资产的占比仍然较小，全球量化CTA基金为主流策略，截至2022年第三季度，量化CTA基金规模占CTA基金总规模90%。

## 1.2 海外 CTA 基金发展
- 1848-1987，期货交易探索期
    - 1848年,谷物商成立芝加哥交易所
    - 1949年，美国证券经纪人Richard Donchian设立第一个公开发售的期货基金，标志着CTA基金的诞生
    - 1971年，管理期货行业协会（Managed Futures Association）的建立， 标志着CTA正式成为业界所接受的一种投资策略
- 1987-2000，CTA基金高速发展
- 2001-2012，CTA基金进入成熟的阶段
- 2013-2019，CTA基金进入瓶颈
- 2020-2022，CTA基金再次爆发
    - 2020年，全球疫情的爆发带来了阶段性行情
    - 2021-2022年，全球性通货膨胀下，CTA基金，尤其是趋势策略爆发式增长

截至 2022 年，根据巴克莱基金数据，全球 CTA 基金规模已达到 4050 亿美元左右。

# 二、趋势性CTA策略

## 2.1 趋势择时策略介绍
- 趋势型指标是投资者运用最多，也最容易在市场中获利的方法。市场中最为著名的格言:“让利润充分增长，限制损失”，是趋势型指标的真实反映。
- 趋势型指标通常利用**两根线的交叉**作为交易信号，并以此作为买卖时点的判断。
- 优点：
     - （1）确定性高；  
     - （2）准确时可以获得高利润。
- 缺点：
     - （1）多数趋势指标滞后于价格走势，只能吃到‘鱼身’；
     - （2）冲击成本巨大；
     - （3）胜率低。

## 2.2 均线策略
### MA（移动平均）
移动平均分析是利用统计学上移动平均的原理，对每天的股价或成交数据进行平均化处理，以消除偶然变动，减弱季节和循环变动的影响。  
移动平均的计算方法有多种，最常用的是算术移动平均，又称为简单移动平均（SMA），其计算公式为：  
$$  {SMA}(N)_{t}=\frac{1}{N} \sum_{i=0}^{N-1} P_{t-i} $$
其中，N为移动平均期间，$ P_{t-i} $为第(t-i)天的收盘价或股价指数。

**择时策略**：
- 葛南维移动平均线八大法则:四条用来研判买进时机，四条用来研判卖出时机。
- 交叉择时法则：
    - 即当一条短期均线从下向上穿过长期均线时，形成所谓**金叉**，此时应该做多；
    - 当一长期均线从上向下穿过短期均线时，形成所谓**死叉**，此时应该做空或空仓。

**策略测试**：
- 测试参数包括计算短期均线天数S和长期均线天数L；在每个测试期间内，S以2天为间隔，测试范围从2天到20天。
- 测试中采用遍历的搜索方法，分别计算不同参数匹配下的择时交易情况。
- 测试结果：
    - 以4日为短期均线，40日为长期均线进行交叉择时效果相对较好。
    - 从测试情况来看，MA指标适合长线择时。在不考虑交易成本的情况下，交叉择时交易法则能获得不错的收益表现。

#### 双均线择时策略-代码实现

提示：需要提前安装akshare、pandas_ta等包，关于akshare的介绍：https://akshare.akfamily.xyz/introduction.html

```python
import warnings
warnings.filterwarnings('ignore')
import akshare as ak
import pandas as pd
import numpy as np
import pandas_ta as ta
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  #中文字体
plt.rcParams['axes.unicode_minus'] = False  #解决坐标轴负数的负号显示问题

future_df = ak.futures_zh_daily_sina(symbol="RB0") #获取螺纹钢连续合约数据
print(future_df)
start_date = pd.to_datetime('2014-01-01')
end_date = pd.to_datetime('2022-12-31')
price_df=future_df[['date','open','high','low','close']]
price_df['date'] = pd.to_datetime(price_df['date']).dt.date
price_df = price_df[(price_df['date']>=start_date) & (price_df['date']<=end_date)]
price_df = price_df.sort_values('date').set_index('date')

# 计算每日的收益率
price_df['returns'] = price_df['close'].pct_change().shift(-1).fillna(0)
display(price_df.head())

#双均线择时
days_s = 4  #均线参数可修改
days_l = 40
price_df['ma_s'] = ta.sma(price_df['close'], length=days_s)
price_df['ma_l'] = ta.sma(price_df['close'], length=days_l)
# 择时信号：当日短期均线值大于长期均线值则开仓，否则清仓
timing_df = pd.DataFrame()
timing_df['择时'] = (price_df['ma_s']>price_df['ma_l']) * 1.
timing_df['不择时'] = 1.
# 计算择时后的每日收益率
timing_ret = timing_df.mul(price_df['returns'], axis=0)
timing_ret['超额收益'] = (1+timing_ret['择时']).div(1+timing_ret['不择时'], axis=0) - 1.
display(timing_ret.head())
# 计算累计收益率
cumul_ret = (1 + timing_ret.fillna(0)).cumprod() - 1.
cumul_ret.plot(figsize=(10, 6), title='双均线择时_累计收益率')
```

### MACD（指数平滑异同移动平均线）
MACD 指标是根据均线的构造原理，通过分析短期(通常为12日)指数移动平均线与长期(通常为26日)指数移动平均线之间的聚合与分离状况，对买进、卖出时机做出判断的技术指标。

**MACD的计算：**
1. 计算短期(S日)指数移动平均线和长期(L日)指数移动平均线EMA1、EMA2；
2. 计算离差值DIFF=EMA1-EMA2，其反映价格是多头排列还是空头排列；
3. 计算DIF的N日指数移动平均线，即 DEA；
4. 计算 MACD=2*(DIF-DEA)。  

**MACD图示:**

<img src="D:/研一/量金/MACD图示.png" width="60%">

如图，白线是12日与26日均线的差离值（DIFF），黄线是DIF的9日移动平均线（DEA），红绿区块柱形图（MACD）是两者的差，可以表示股价涨跌的内在动力。  
位于零轴上方的MACD红色柱线越长，说明股价上涨动能越强。位于零轴下方的MACD绿色柱线越长，说明股价的下跌动能越强。  
在 MACD 的计算和测试中，需要设定的参数主要包括短期均线和长期均线的计算天数 S、L，以及 DEA 的计算天数 M。

**择时策略:**
1. DIFF、DEA均为正，DIFF向上突破DEA，买入信号；
2. DIFF、DEA均为负，DIFF向下突破DEA，卖出信号；
3. DEA与K线发生背离，行情反转信号；
4. 分析MACD柱状线，由红变绿，卖出信号；由绿变红，买入信号。

**案例：MACD柱线和股价顶背离**

<img src="D:/研一/量金/顶背离.png" width="80%">

在股价持续上涨、连创新高的同时，如果MACD柱线没有创新高，反而出现一顶比一顶低的走势，二者就构成了顶背离形态。这样的形态说明虽然股价持续上涨，但是其上涨的内在动能越来越弱，未来股价有见顶下跌的风险。

## 2.3 通道类策略
通道类策略的核心思想是：股票或其他金融资产的价格往往在某个价格区间内波动，而这个区间就被称为“通道”。通道择时策略会设定一个上轨和一个下轨，这两条轨道定义了价格的波动范围。当价格突破这个范围时，可能意味着一个买入或卖出的信号。
### 克罗均线交易系统
斯坦利·克罗是全球顶级的期货投资专家，在33年的职业生涯中，不但赢得了丰厚的回报，也积累了丰富的经验。《克罗谈投资策略》、《期货交易策略》等著作为后人留下了宝贵的精神财富。在这些著作中零星地渗透着克罗的交易思想，如KISS原则（Keep it Simple,Stupid）。  
克罗的交易系统同样基于简单的均线进行交易，并遵循顺势原则。

**克罗均线交易系统构建**
1. 均线系统确定
    - 长期均线组：回溯期分别为10天、20天、50天的长期简单移动平均线。
    - 短期均线组：回溯期分别为4天、9天、18天的短期简单移动平均线。
2. 买入信号
    - 收盘价大于所有长期均线组，并且长期均线组多头排列（MA10>MA20>MA50）；收盘价大于所有短期均线组，并且短期均线组多头排列（MA4>MA9>MA18）。
    - 以上两个信号出现一个即可做多。

3. 卖出信号
    - 收盘价小于所有长期均线组，并且长期均线组空头排列（MA10<MA20<MA50）；收盘价小于所有短期均线组，并且短期均线组空头排列（MA4<MA9<MA18）。
    - 以上两个信号出现一个即可做空。

### 火车轨交易系统
该系统长短皆宜，特别对于短周期交易的盘整行情的过滤有着十分明显的效果。该策略从本质上来说从属于均线系统，但有别于均线系统。

**策略原理**
- 火车轨上轨：一段周期内高点的移动平均值
- 火车轨下轨：一段周期内低点的移动平均值

**交易规则**
1. 买入规则：前一日收盘价高于火车轨上轨，下一日开盘做多；
2. 卖出规则：前一日收盘价低于火车轨下轨，下一日开盘做空
3. 追踪止盈规则：当盈利超过一定幅度时，盈利回撤一定百分比追踪止盈。

### 布林通道
由约翰·布林在1980年代初期发明。布林通道基于统计学中的标准差原理，主要用于衡量价格的高低和波动性。其包含3条线：
- 中轨：通常是20周期的简单移动平均线（SMA），周期值可以调整
- 上轨：通常设定为中轨之上的两个标准差（可以调整）。计算公式为：上轨 = 中轨 + 2 * 标准差
- 下轨：通常设定为中轨之下的两个标准差（可以调整）。计算公式为：下轨 = 中轨 - 2 * 标准差

#### 布林通道策略-代码实现

择时信号为：价格突破通道上轨时开仓，价格跌破通道下轨时清仓，价格在通道内时维持先前的仓位。

```python
#布林通道策略
future_df = ak.futures_zh_daily_sina(symbol="RB0")
start_date = pd.to_datetime('2014-01-01')
end_date = pd.to_datetime('2022-12-31')
price_df=future_df[['date','open','high','low','close']]
price_df['date'] = pd.to_datetime(price_df['date']).dt.date
price_df = price_df[(price_df['date']>=start_date) & (price_df['date']<=end_date)]
price_df = price_df.sort_values('date').set_index('date')
price_df['returns'] = price_df['close'].pct_change().shift(-1).fillna(0)

# 计算布林带（Bollinger Bands）
length = 20  # 均线的周期
std = 2.0  # 通道的宽度设定为几倍标准差
mamode = 'sma'  # 均线的计算方式
b_channel = ta.bbands(price_df['close'], length=length, std=std, mamode=mamode)
display(b_channel[21:31])

# 可视化通道
up = 'BBU_20_2.0'
down = 'BBL_20_2.0'
middle = 'BBM_20_2.0'
plt.figure()
b_channel[[up, down, middle]].plot(figsize=(10, 6))
plt.title('布林通道')
plt.show()

# 计算择时信号:价格突破通道上轨时开仓,价格跌破通道下轨时清仓,价格在通道内时维持先前的仓位
timing_df = pd.DataFrame()
timing_df['择时'] = (price_df['close']>=b_channel[up]) * 1. + (price_df['close']<=b_channel[down]) * -1.
timing_df = timing_df.replace(0, np.nan)  # 先将0替换为NA
timing_df = timing_df.fillna(method='ffill').fillna(0)  # 使用前值填充NA
timing_df[timing_df<0] = 0
timing_df['不择时'] = 1.

# 计算择时和不择时的每日收益率
timing_ret = timing_df.mul(price_df['returns'], axis=0)
timing_ret['超额收益'] = (1+timing_ret['择时']).div(1+timing_ret['不择时'], axis=0) - 1.

# 计算累计收益率
cumul_ret = (1 + timing_ret.fillna(0)).cumprod() - 1.
cumul_ret.plot(figsize=(10, 6), title='布林通道择时')
```

### 凯特纳通道
凯特纳通道交易系统是由技术分析专家Chester Keltner在上世纪50年代开发；  
凯特纳通道可以让交易员很快地观察到股价的趋势是向上还是向下，或横盘走，也可以判断可能的支撑或压力区；  
和布林通道相比，凯特纳通道更加平滑。

**通道确定**
- 中轨：N日移动平均线
- 上轨：中轨+N倍ATR
- 下轨：中轨-N倍ATR  
其中ATR为平均真实波幅，取以下3个值的最大值：

**交易规则**
- 突破上轨，买入开仓做多；
- 突破下轨，卖出开仓做空；
- 多头持仓，前一根K线的收盘价跌破中轨平仓；
- 空头持仓，前一根K线的收盘价上穿中轨平仓。

#### 凯特纳通道策略-代码实现

```python
#凯特纳通道策略代码实现(与布林通道选择的合约相同)

# 计算凯特纳通道（Keltner Channels）
length = 20  # 均线的周期
scalar = 2  # 通道的宽度设定为几倍ATR
mamode = 'ema'  # 均线的计算方式
k_channel = ta.kc(price_df['high'], price_df['low'], price_df['close'], length=length, scalar=scalar, mamode=mamode)
display(k_channel[21:31])

#可视化通道
up = 'KCUe_20_2.0'
down = 'KCLe_20_2.0'
middle = 'KCBe_20_2.0'
plt.figure()
k_channel[[up, down, middle]].plot(figsize=(10, 6))
plt.title('凯特纳通道')
plt.show()

# 计算择时信号:价格突破通道上轨时开仓,价格跌破通道下轨时清仓,价格在通道内时维持先前的仓位
timing_df = pd.DataFrame()
timing_df['择时'] = (price_df['close']>=k_channel[up]) * 1. + (price_df['close']<=k_channel[down]) * -1.
timing_df = timing_df.replace(0, np.nan)  # 先将0替换为NA
timing_df = timing_df.fillna(method='ffill').fillna(0)  # 使用前值填充NA
timing_df[timing_df<0] = 0
timing_df['不择时'] = 1.

# 计算择时和不择时的每日收益率
timing_ret = timing_df.mul(price_df['returns'], axis=0)
timing_ret['超额收益'] = (1+timing_ret['择时']).div(1+timing_ret['不择时'], axis=0) - 1.
# 计算累计收益率
cumul_ret = (1 + timing_ret.fillna(0)).cumprod() - 1.
cumul_ret.plot(figsize=(10, 6), title='凯特纳通道择时')
```

## 3.1 跨期套利策略
### 跨期套利理论介绍
- 当两个不同到期月份合约之间的价差偏离其合理区间时，可以通过在期货市场同时买入低估值合约和卖出高估值合约，在价差回归后进行反向平仓的方式来进行跨期套利交易。
- 跨期套利的操作重点在于判断不同到期月合约的价差将来是扩大还是缩小。
- **均衡价差**：
    - 从理论上看，不同合约间价差会收敛并趋向于均衡价差，这就是同一标的指数的不同交割月股指期货合约之间存在着的一种平价关系；
    - 远月合约的价值应该是近月合约价值按照远期利率进行复利后加上一个均衡价差，其理论值可以参考不同合约间的平价计量模型：
    
<img src="D:/研一/量金/均衡价差.png" width="50%">

- **无套利区间**:
    - 无套利价差区间是指综合考虑了融资成本、交易成本等方面的因素后，相关期货合约之间价差的合理区间；
    - 只有当价差收益大于交易成本时才可以进行跨期套利,当价差在该区间内变动时，不存在套利机会；
    - 一旦价差大于无套利价差区间上边界，就可以采用**卖出套利策略**;一旦价差小于无套利价差区间下边界，就可以采用**买进套利策略**，直至合约间价差又回归到无套利区间内，跨期套利合约就可以适时**平仓**。
    
### 应用-牛市套利   
- 对于大多数商品期货来说，当市场是牛市（看涨的）时，较近月份的合约价格上涨幅度往往要大于较远期合约价格的上涨幅度。因而，远期合约价格与较近月份合约价格之间的价差往往会缩小。
- 在这种情况下，买入较近月份的合约同时卖出远期月份的合约进行套利盈利的可能性比较大。
- 套利收益=仓位*((远期月份卖出价-近期月份买进价)-(远期月份买进价一近期月份卖出价))

- 所谓熊市套利，是指卖出较近月份的合约同时买入远期月份的合约，这正好与牛市套利策略相反。
- 之所以称熊市套利，是因为一般认为当市场是熊市时，较近月份合约价格下降的幅度通常大于较远期合约价格的下降幅度，远期合约价格与较近月份合约价格之间的价差往往会扩大。
- 套利收益=仓位*((远期月份买进价-近期月份卖出价)-(远期月份卖出价一近期月份买进价))

## 3.2 跨品种套利策略

##  跨品种套利介绍
- 跨品种套利指利用两种或多种不同的，但相互关联的商品之间的合约价格差异进行套利交易，即买入某一交割月份某种商品合约，同时卖出另一相同交割月份相互关联的商品合约。
- 这几种商品在用途上存在着替代关系，或者这几种商品受到同一种供求因素的影响，它们在期货市场上表现为合约价格的变动方向相向或相反。
- 跨期套利主要是正向套利与反向套利，而跨品种的套利更多的强调**强弱对冲**。

- **统计套利步骤**：

2. 进行ADF检验平稳性和协整检验

5. 历史数据回测

#### 统计套利策略-代码实现

```python
rb_df = ak.futures_zh_daily_sina(symbol="RB2410") # 获取螺纹钢2410合约数据
hc_df = ak.futures_zh_daily_sina(symbol="HC2410") # 获取热卷2410合约数据
day=rb_df.index
rb=rb_df['close']
hc=hc_df['close']
fig = plt.figure(figsize=(6, 4))
plt.plot(day,rb,color='blue',label='螺纹钢2410')
plt.plot(day,hc,color='green',label='热卷2410')
plt.legend()
plt.show()

def strategy(price_A,price_B):
    spread = price_A - price_B
    mspread = spread - np.mean(spread)
    sigma = np.std(mspread)
    open = 1.5 * sigma # 参数可调整
    stop = 2 * sigma
    profit_list = []
    hold = False
    hold_price_A = 0
    hold_price_B = 0
    hold_state = 0  # 1 (A:long B:short) -1 (A:short B:long)
    profit_sum = 0
    
    for i in range(len(price_A)):
        if hold == False:
            if mspread[i] >= open:
                hold_price_A = price_A[i]
                hold_price_B = price_B[i]
                hold_state = -1
                hold = True
            if mspread[i] <= -open:
                hold_price_A = price_A[i]
                hold_price_B = price_B[i]
                hold_state = 1
                hold = True
            profit_list.append(profit_sum)
            continue
            
        else:
            if mspread[i] >= stop and hold_state == -1:
                profit = (hold_price_A - price_A[i]) + (price_B[i] - hold_price_B)
                profit_sum += profit
                hold_state = 0
                hold = False
            if mspread[i] <= -stop and hold_state == 1:
                profit = (price_A[i] - hold_price_A) + (hold_price_B - price_B[i])
                profit_sum += profit
                hold_state = 0
                hold = False
            if mspread[i] <= 0 and hold_state == -1:
                profit = (hold_price_A - price_A[i]) + (price_B[i] - hold_price_B)
                profit_sum += profit
                hold_state = 0
                hold = False
            if mspread[i] >= 0 and hold_state == 1:
                profit = (price_A[i] - hold_price_A) + (hold_price_B - price_B[i])
                profit_sum += profit
                hold_state = 0
                hold = False
            profit_list.append(profit_sum)
            continue
        
    return profit_list

price_A = rb_df['close'].values
price_B = hc_df['close'].values
profit_list=strategy(price_A,price_B)
fig = plt.figure()
plt.plot(range(len(profit_list)), profit_list)
plt.title('跨品种套利回测结果')
plt.show()     
```
