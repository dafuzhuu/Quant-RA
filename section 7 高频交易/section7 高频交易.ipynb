# 2.高频交易的定义与特征

## 2.1高频交易的定义

高频交易 (HFT) 指的是利用复杂的技术工具和计算机算法来快速进行证券交易。这些系统能够在极短时间内处理大量数据并执行订单，通过微小的价格波动来实现利润。高频交易的特点是高订单比率、极短的持仓周期以及对速度和先进基础设施的依赖，从而在市场中获得竞争优势。

- **速度：** 高频交易以微秒甚至纳秒为时间单位进行操作。
- **交易量：** 高频交易策略每天执行大量交易。
- **算法：** 专有算法根据市场数据做出交易决策。
- **基础设施：** 先进的硬件和网络连接对于降低延迟至关重要。

## 2.2高频交易的关键特征

1. 利用超高速的复杂计算机系统进行下单
   - 高频交易依赖于最先进的计算机系统，这些系统能够以前所未有的速度处理数据和执行交易。它们采用强大的处理器、高速内存和优化的软件来减少从接收到市场数据到采取行动之间的延迟
2. 利用co-location和直连交易所的数据通道进行交易
   - 主机托管（co-location）指的是将服务器放到交易所数据中心，从而降低延迟。这也成为交易所一项有利可图的业务，它向高频交易公司收取数百万美元的费用以提供这种“低延迟访问”（low latency access）的特权。
   - 此外，高频交易公司利用直接市场接入 (DMA) 数据通道建立与交易场所的直接连接，绕过中介机构，可以更快地与交易所进行通信
3. 平均每一次持仓的时间非常非常的短
    - 高频交易策略通常仅持有头寸极短时间，往往以秒甚至毫秒为单位计算。其目的是利用短期的价格差异，而非长期的市场趋势
    - 短期持仓减少了对不利价格变动的敞口
    - 短持仓时间符合通过短暂市场机会获利的目标，这是高频交易和中低频交易的明显区别、
4. 能够大量的发送和取消委托订单

**机制：**
- 频繁地下达和撤回买卖订单来影响市场的供需关系和价格走向（订单簿操作）
- 不断调整买卖订单以保持所需的头寸

**技术需求：**
- 需要高效的订单管理系统 (OMS) 来处理高量订单流且延迟最低。
- 算法能根据预设标准迅速决定何时下达或取消订单。

5. 在收盘的时候，几乎是保持空仓的，决不持仓过夜

高频交易公司通常会在交易日结束前清空所有头寸，确保没有隔夜持仓。这一做法消除了隔夜市场风险，例如由于非交易时间的新闻或事件导致的价格跳空。

这样做有两点优势

- 避免了因市场休市时持仓带来的不确定性
- 为第二天的新交易机会释放资金


# 3.高频交易类型

## 3.1订单拆分策略

**市场冲击**指的是由于执行大额交易对证券价格产生的不利影响。当机构投资者下达大量买单或卖单时，这会改变供需平衡，从而导致价格向不利方向移动。

设 $ P_0 $ 为证券的初始价格，预期的执行价格 $ P_{\text{exec}} $ 可表示为：

$$
P_{\text{exec}} = P_0 + \lambda \cdot \text{Order Size}
$$

其中 $ \lambda $ 是市场冲击系数，**Order Size** 为订单量相对于日均交易量的比例。

市场冲击的存在对高频交易者不利，体现在以下几个方面

1. 增加交易成本：买单时价格上升，卖单时价格下降
2. 价格滑点：实际执行价格与预期价格产生偏差
3. 信息泄露：其他市场参与者可能察觉到大额订单的存在，提前交易，加剧价格波动

订单拆分策略为了解决这个问题，使用多种算法把大订单分割成若干个小订单，从而减小订单对市场的影响并降低执行成本。

- 第一代主要考虑如何减小对市场的影响
- 第二代拆分策略则加入了一些反侦测的技术手段
- 第三代拆分策略认为，如果片面地强调订单分割和避免被侦测，就存在无法按时完成交易计划的风险，这样反而会导致交易成本的上升，因此强调利用交易量较大的交易时间完成仓位计划

### 第一代：最小化市场冲击

第一代订单拆分策略的重点是通过逐步执行订单从而最小化即时市场冲击。主要有两种算法主导这一代策略：**时间加权平均价格（TWAP）**和**成交量加权平均价格（VWAP）**。

#### a. 时间加权平均价格（TWAP）

TWAP 将总订单 $Q$ 分为 $N$ 等量交易，并按固定时间间隔 $\Delta t$ 执行。执行计划表示为：

$$
q_i = \frac{Q}{N} \quad \text{对于} \quad i = 1, 2, \dots, N
$$

其中 $q_i$ 是第 $i$ 个时间间隔执行的交易量。

**例子**

假设某投资者计划在一个交易日内购买 $Q = 1,000,000$ 股（6 小时或 360 分钟），并使用 TWAP 以 $N = 60$ 间隔（每 6 分钟执行一次）：

$$
q_i = \frac{1,000,000}{60} \approx 16,667 \text{ 股每次执行}
$$

每6分钟间隔会购买约16,667股，将订单均匀分布于交易日内。

#### b. 成交量加权平均价格（VWAP）

VWAP 将订单执行与随时间变化的实际交易量分布 $V(t)$ 对齐。时间 $t$ 的执行率 $q(t)$ 与观测到的交易量成正比：

$$
q(t) = Q \cdot \frac{V(t)}{\int_{0}^{T} V(s) \, ds}
$$

其中 $V(t)$ 是时间 $t$ 的交易量，$T$ 是总交易时段。

**例子**

假设某投资者计划在一个交易日内卖出 $Q = 500,000$ 股 ABC 公司股票。假设交易量分布在上午10点到下午2点之间较高，占全天交易量的60%。执行计划会分配：

$$
q_{\text{高峰}} = 0.6 \times 500,000 = 300,000 \text{ 股}
$$
$$
q_{\text{非高峰}} = 0.4 \times 500,000 = 200,000 \text{ 股}
$$

这种分配确保订单的大部分在交易高峰时段执行，与市场的自然活动保持一致。


### 第二代：引入防检测技术

第一代策略仅仅解决了交易成本和滑点的问题，信息泄露却依然存在。执行大额订单可能会向其他市场参与者发出信号，导致抢先交易或不利的价格波动。

#### a. 冰山订单

冰山订单在任何时候只显示总订单 $Q$ 的一小部分 $f$。可见部分 $q_{\text{visible}}$ 定义为：

$$
q_{\text{visible}} = f \cdot Q
$$

在可见部分执行后，下一批订单将被显示：

$$
q_{i} = f \cdot Q \quad \text{对于} \quad i = 1, 2, \dots, \frac{1}{f}
$$

**例子**

某投资者计划购买 $Q = 2,000,000$ 股，并设定冰山系数 $f = 0.01$：

$$
q_{\text{visible}} = 0.01 \times 2,000,000 = 20,000 \text{ 股}
$$

每次执行20,000股后，会再显示20,000股，保持小额订单的假象。

#### b. 订单量和时间间隔的随机化

为防止模式检测，订单量 $q_i$ 和执行间隔 $\Delta t_i$ 进行了随机化。设 $q_i$ 遵循概率分布 $P_q$，执行间隔 $\Delta t_i$ 遵循概率分布 $P_t$：

$$
q_i \sim P_q, \quad \Delta t_i \sim P_t
$$

常用的分布包括指定范围内的均匀分布或正态分布。

**例子**

某投资者计划购买 $Q = 500,000$ 股。算法从10,000到30,000股的均匀分布中随机选择 $q_i$，并从平均为5分钟的指数分布中选择时间间隔 $\Delta t_i$。这种随机化方法使其他市场参与者难以识别隐藏的大额订单。


### 第三代：平衡执行和防检测

如果一味地降低交易成本，可能无法完成当日的交易任务，导致持有头寸过夜，遭受更高的风险和成本。因此，我们需要找到一种平衡执行效率与市场冲击之间的平衡点。

第三代策略超越了单纯的冲击最小化和防检测，采用**自适应算法**平衡隐蔽性和及时执行。这一代策略解决了订单延迟完成的风险，避免因市场波动增加成本。

#### a. 基于市场条件的自适应算法

自适应算法使用实时数据动态调整执行率。常见方法使用**随机控制**，其中执行率 $q(t)$ 是可观测市场变量 $X(t)$ 的函数：

$$
q(t) = \phi(X(t))
$$

其中 $\phi$ 是通过优化原则（如最小化预期成本）得出的策略函数：

$$
\min_{\phi} \mathbb{E} \left[ \int_{0}^{T} (P(t) \cdot q(t) + \lambda \cdot q(t)^2) \, dt \right]
$$

这里，$P(t)$ 是时间 $t$ 的资产价格，$\lambda$ 是控制执行率的参数，旨在控制市场冲击。

**例子**

在高流动性时段（如交易量增加时），算法会设置较高的执行率 $q(t)$，以利用较低的市场冲击。相反，在低流动性时段，$q(t)$ 被降低以避免显著的价格波动。

#### b. 智能订单路由（SOR）

SOR 优化跨多个交易场所的订单执行，以获得最佳价格。这涉及解决一个优化问题，目标是最小化执行成本 $C$：

$$
\min_{\{q_i\}} C = \sum_{i=1}^{M} \left( P_i \cdot q_i + \eta_i \cdot \frac{q_i^2}{V_i} \right)
$$

约束条件为：
$$
\sum_{i=1}^{M} q_i = Q, \quad q_i \geq 0 \quad \forall i
$$

其中：
- $M$ 是交易场所的数量，
- $P_i$ 是第 $i$ 个场所的价格，
- $\eta_i$ 是第 $i$ 个场所的市场冲击系数，
- $V_i$ 是第 $i$ 个

场所的平均交易量，
- $q_i$ 是分配到第 $i$ 个场所的订单量。

**例子**

假设某投资者需要在 $M = 3$ 个交易场所卖出 $Q = 1,000,000$ 股，参数如下：

| 场所 | $P_i$ | $\eta_i$ | $V_i$ |
|-------|---------|------------|---------|
| 1     | 50.00   | 0.5        | 5,000,000 |
| 2     | 50.10   | 0.3        | 3,000,000 |
| 3     | 49.90   | 0.4        | 4,000,000 |

优化问题变为：

$$
\min_{q_1, q_2, q_3} \left( 50.00q_1 + 0.5 \frac{q_1^2}{5,000,000} + 50.10q_2 + 0.3 \frac{q_2^2}{3,000,000} + 49.90q_3 + 0.4 \frac{q_3^2}{4,000,000} \right)
$$

约束条件为：
$$
q_1 + q_2 + q_3 = 1,000,000 \quad \text{且} \quad q_i \geq 0
$$

通过求解这一二次优化问题，可以得到最优的 $q_i$ 分配策略，最小化总执行成本。

#### c. 交易量预测集成

集成预测分析涉及使用时间序列模型（如**自回归积分滑动平均（ARIMA）**）或机器学习算法预测未来的交易量 $V(t)$。预测的交易量用于指导自适应执行策略：

$$
\hat{V}(t+\Delta t) = \text{ForecastModel}(V(t))
$$

然后，根据预测的 $\hat{V}(t+\Delta t)$ 调整执行率 $q(t)$，以优化订单执行。

**例子**

某算法预测在公司业绩公告前交易量将大幅上升。使用 ARIMA 模型预测 $V(t+\Delta t)$ 增加30%。因此，算法会在高交易量时段加快订单执行，以确保及时完成，同时保持较低的市场冲击。

## 3.2流动性回扣交易
